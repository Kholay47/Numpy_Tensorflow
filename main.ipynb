## Importing Necessary Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

## Generating Random Dataset Using Numpy

```python
observations = 1000

xs = np.random.uniform(low = -10, high = 10, size = (observations,1))
zs = np.random.uniform(-10,10,(observations,1))
```

```python
generated_inputs = np.column_stack((xs,zs))
```

```python
noise = np.random.uniform(-1,1,(observations,1))
```

```python
generated_targets = 2*xs - 3*zs + 5 +noise
```

```python
np.savez('data_set', inputs = generated_inputs, targets = generated_targets)
```

## Solving Using TensorFlow

```python
training_data = np.load('data_set.npz')
```

```python
input_size = 2
output_size = 1

model = tf.keras.Sequential([tf.keras.layers.Dense(output_size)])


model.compile(optimizer='sgd', loss='mean_squared_error')

model.fit(training_data['inputs'], training_data['targets'],epochs = 100)
```

## Extracting Weight And Bias

```python
model.layers[0].get_weights()
```

```python
weights = model.layers[0].get_weights()[0]
weights
```

```python
bias = model.layers[0].get_weights()[1]
bias
```

## Predictions

```python
model.predict_on_batch(training_data['inputs'])
```

## Plotting

```python
plt.plot(np.squeeze(model.predict_on_batch(training_data['inputs'])), np.squeeze(training_data['targets']))
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show
```
